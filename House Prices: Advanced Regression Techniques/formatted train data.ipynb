```python
# Import libraries
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns

# Import the test dataset
train_df=pd.read_csv('train.csv')
train_df.tail()
```

```python
## check for null values
train_df.isnull().sum()
```

```python
new_train_df=train_df
```

```python
new_train_df.tail()
```

```python
new_train_df.info()
```

```python
# Drop the columns having more than 50% null values
new_train_df.drop(['Id','Alley','GarageYrBlt','PoolQC','Fence','MiscFeature','FireplaceQu'],axis=1,inplace=True)
```

```python
new_train_df.shape
```

```python
new_train_df['MSZoning']=new_train_df['MSZoning'].fillna(new_train_df['MSZoning'].mode()[0])
new_train_df['LotFrontage']=new_train_df['LotFrontage'].fillna(new_train_df['LotFrontage'].mean())
new_train_df['BsmtCond']=new_train_df['BsmtCond'].fillna(new_train_df['BsmtCond'].mode()[0])
new_train_df['BsmtQual']=new_train_df['BsmtQual'].fillna(new_train_df['BsmtQual'].mode()[0])
new_train_df['BsmtExposure']=new_train_df['BsmtExposure'].fillna(new_train_df['BsmtExposure'].mode()[0])
new_train_df['BsmtFinType2']=new_train_df['BsmtFinType2'].fillna(new_train_df['BsmtFinType2'].mode()[0])
```

```python
new_train_df['GarageType']=new_train_df['GarageType'].fillna(new_train_df['GarageType'].mode()[0])
new_train_df['GarageFinish']=new_train_df['GarageFinish'].fillna(new_train_df['GarageFinish'].mode()[0])
new_train_df['GarageQual']=new_train_df['GarageQual'].fillna(new_train_df['GarageQual'].mode()[0])
new_train_df['GarageCond']=new_train_df['GarageCond'].fillna(new_train_df['GarageCond'].mode()[0])
new_train_df['MasVnrType']=new_train_df['MasVnrType'].fillna(new_train_df['MasVnrType'].mode()[0])
new_train_df['MasVnrArea']=new_train_df['MasVnrArea'].fillna(new_train_df['MasVnrArea'].mode()[0])
```

```python
new_train_df.loc[:, new_train_df.isnull().any()].head(10)
```

```python
new_train_df['BsmtFinType1']=new_train_df['BsmtFinType1'].fillna(new_train_df['BsmtFinType1'].mode()[0])
new_train_df['Electrical']=new_train_df['Electrical'].fillna(new_train_df['Electrical'].mode()[0])
```

```python
new_train_df.loc[:, new_train_df.isnull().any()].head(3)
```

```python
new_train_df.head(10)
```

```python
new_train_df.to_csv('final_train_data.csv',index=False)
```
